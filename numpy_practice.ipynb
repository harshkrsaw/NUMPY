```python
a = np.random.randint(1,100,24).reshape(6,4)
```

```python
a
```

```python
#A in greater than 50
a > 50
```

```python
a[a>50]
```

```python
#find all number greater than 50 and even
a[(a>50) & (a%2 == 0)]
```

```python
#find all the number not divisible by 7
a[a%7 != 0]
```

```python
a[~(a%7 == 0)]
```

 ## Broadcasting

```python
a = np.arange(12).reshape(4,3)
a
```

```python
b = np.arange(2).reshape(2)
b
```

```python
a + b
```

```python
# More examples

a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)
print(b)

print(a+b)
#broadcasting is possible
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3)

print(a)
print(b)

print(a+b)
#broadcasting is not possible
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)

print(a+b)
#broadcasting is possible
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)

print(a)
print(b)

print(a + b)
#broadcasting is possible
```

```python
a = np.array([1])
# shape -> (1,1)
b = np.arange(4).reshape(2,2)
# shape -> (2,2)

print(a)
print(b)

print(a+b)
#broadcasting is possible
```

```python
a = np.array([1])
# shape -> (1,1)
b = np.arange(4).reshape(2,2)
# shape -> (2,2)

print(a)
print(b)

print(a+b)
#broadcasting is possible
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

print(a)
print(b)

print(a+b)
#broadcasting is not possible
```
